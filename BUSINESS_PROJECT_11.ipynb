# TASK 10 - Financial ratios (0.5 points)
**DIY** - Use SABI database at the UPNA Library and Python to calculate your industry's financial ratios:

```python
# Liquidity ratios
currentassets = 5090587
currentliabilities = 4284412
stocks = 2549139

CR = currentassets/currentliabilities
ATR = (currentassets-stocks)/currentliabilities
print("CR: ", CR,", ATR: ", ATR)

# Solvency ratios
noncurrentliabilities = 1548419
Debt = currentliabilities + noncurrentliabilities
Assets = 6376609
Operating_PL = -135232
Financial_Expenses = 10164

DAR = Debt/Assets
TIER = Operating_PL/Financial_Expenses
print("DAR: ", DAR,", TIER: ", TIER)


# Profitability ratios
NetIncome = -170161
Equity = Assets - Debt
ROA = NetIncome/Assets
ROE = NetIncome/Equity

print("ROA: ", ROA,", ROE: ", ROE)
```

**DIY** - Explain your industry's financial ratios:

As we have calculated the ratios, we can see from the Current Ratio that the number falls into a category where it is alarming due to being less than 1.5. This means that the company would have troubles in fulfilling short term liabilities due to the low liquidity. Synonymously, we can see that the ATR number is also low. Thus, we can conclude that the company's liquidity is low.

In terms of the solvency, a DAR number of around 0.9 means that the company still has more assets than debts, though the margin might be a bit small. The negative TIER figure also shows that the company is suffering from loss.
