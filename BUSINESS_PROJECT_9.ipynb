# TASK 8 - Break-even point (0.5 points)
**DIY** - Use Matplotlib from Python to plot a graph that contains the two Operating P/L functions generated by each of your business's technologies:

```python
import matplotlib.pyplot as plt
import numpy
x = numpy.linspace(0,10,100)
y1 = (2500)*x - (33938/2)
y2 = (2800)*x - (25560/2)
plt.plot(x,y1,'red')
plt.plot(x,y2,'green')
plt.axhline(0)
plt.axvline(0)
plt.text(5,-5,'$q_{B}$')
plt.xlabel("Units of product sold")
plt.ylabel('Operating P/L')
```

**DIY** - Explain the **break-even point** of each technology: The above graph represents the EBIT of both of 
our production methods ( CNC (green) and AM (red)). 

CNC's production is more expensive, consequently, we will sell products manufactured with this technology at a higher prize. Therefore, the difference between the prize and the variable operating costs is bigger and the slope is greater. So the break even point is reached sooner. Another factor that infuences could be the different values of F (fixed operating expenses), it is higher for 3D-printing. Profit is made when more than 4.5 CNC products are sold and more than 6.7 3D-printed products are sold. Those are the break even points.
