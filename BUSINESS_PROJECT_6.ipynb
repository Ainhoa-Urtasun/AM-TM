**DIY** - Use Python to calculate the Net Present Value (NPV) of your business's investment in AM and the NPV of your business's investment in CNC:

# TASK 8 - Break-even point (0.5 points)
**DIY** - Use Matplotlib from Python to plot a graph that contains the two Operating P/L functions generated by each of your business's technologies:

```python
import matplotlib.pyplot as plt
import numpy
x = numpy.linspace(0,100,100)
AM = 300*x - 15000.0
CNC = 599*x - 15000.0
plt.plot(x,AM,'red')
plt.plot(x,CNC,'green')
plt.axhline(0)
plt.axvline(0)
plt.text(25.04,0,'$q_{B}$')
plt.text(50,0,'$q_{B}$')
plt.xlabel("Units of product sold")
plt.ylabel('Operating P/L')
```

**DIY** - Explain the **break-even point** of each technology:

The break-even point of AM is 50 units. Which means that with a fixed operating cost of 15000, with a unit variable operating cost of 200 and selling each piece at a price of 500; the pieces that our company needs to sell in order to earn money is 50 units. At that point the cost are paid by the revenue and the business starts having profits.

The same happens with CNC machinery. The break-even point is 25.04 units, therefore with a fixed operating cost of 15000, a unit variable operating cost of 400 and a unit selling price of 999; the company needs to sell 25.04 units in order to stop having losses and start having profits.
